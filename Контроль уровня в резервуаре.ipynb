```python
import pandas as pd
```

## Моделируем данные уровня жидкости

```python
df = pd.DataFrame(index = pd.date_range(start='1/1/2018',end = '1/2/2018', freq='30Min'))
df.head()
```

```python
df['Уровень'] = 50
df.head()
for ind in df.index:
    ur = df.loc[ind, 'Уровень'] 
    if ind > pd.to_datetime('2018-01-01 10:30:00'):
        df.loc[ind, ['Уровень']] = int(ur - 1)
df
```
